```python

user_query="i am an indian and i live in india"
```

```python

document="india is a country for the indians and for eveyone"
```

```python
# conver it into thr vectors first 
from collections import Counter 
import math 
```

```python
user_query = user_query.lower().split(" ")
user_query
```

```python
document = document.lower().split(" ")
document
```

```python
user_query_counter = Counter(user_query)
user_query_counter
```

```python
document_counter = Counter(document)
document_counter
```

```python
# lets get only the vectors from the document and user query

list=[]
for tokens in user_query_counter.keys():
    list.append(user_query_counter[tokens])
list
```

```python
for i in user_query_counter.keys():
    print(i)

print()

for i in user_query_counter.keys():
    print(user_query_counter[i])
```

```python
list=[]
for tokens in document_counter.keys():
    list.append(document_counter[tokens])
list
```

```python
for tokens in document_counter.keys() & user_query_counter.keys():
    
    print(tokens)
```

```python
my_list = []
for tokens in user_query_counter.keys() & document_counter.keys():
    my_list.append(user_query_counter[tokens]*document_counter[tokens])

my_list
```

```python
dot_product = sum(my_list)
print(dot_product)
```

```python
query_magnitude = math.sqrt(sum([user_query_counter[tokens]**2 for tokens in user_query_counter.keys()]))
print(query_magnitude)
```

```python
document_magnitude = math.sqrt(sum([document_counter[tokens]**2 for tokens in document_counter.keys()]))
print(query_magnitude)
```

```python
similarity = dot_product / (query_magnitude * document_magnitude)
print(similarity)
```
